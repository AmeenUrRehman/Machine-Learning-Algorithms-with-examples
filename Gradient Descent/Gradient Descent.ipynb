```python
#import libraries
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gradient_descent(x,y):
    
    # Let Slope  = m and Intercept = C in  y  = mx + C and intialise them
    
    m = C = 0
    
    n = len(x)
    
    learning_rate = 0.01
    
#     plt.scatter( x , y , color = 'red' , marker = '+' , linewidth='4')
    
    for i in range(100):
        
        #Let Y predict be Yp
        Y_predicted = m * x + C
        
        cost_function = (1/n) * sum([val**2 for val in (y - Y_predicted)]) 
        
        # Print line for visualization
        plt.plot( x , Y_predicted , color = 'blue')
        plt.xlabel("x-axis")
        plt.ylabel("y-axis")
        
        # Derivative of Yp for m denoted with mder
        mder = - (2/n) * sum ( x * ( y - Y_predicted ) )
        
        # Derivative of Yp for C denoted with Cder
        Cder = - (2/n) * sum ( y - Y_predicted )
        
        m =m - learning_rate * mder
        
        C = C - learning_rate * Cder    
        
        print( " Slope {} , Intercept {} , cost_function {}".format(m,C,cost_function))
        
    
```

```python
x = np.array([1,2,4,5,6])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
